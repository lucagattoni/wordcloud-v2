```python
import wikipedia
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
# parameters

debug = True

additional_stopwords = {'envelope', 'virus'}
```

```python
all_stopwords = additional_stopwords.union(STOPWORDS)

if debug:
    print(all_stopwords)
```

```python
article = wikipedia.page("Coronavirus")
if debug:
    print (article.url)
```

```python
wc = WordCloud(width=1000, height=1000, background_color="white", stopwords=all_stopwords).generate(article.content)
```

```python
plt.figure(figsize=(18, 18))
plt.axis("off")
plt.imshow(wc);
```
