```python
import wikipedia
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
# parameters

debug = True

page = "Cybersecurity"
additional_stopwords = {'even', 'post', 'considered', 'using', 'making', 'part', 'different', 'possible', 'designed', 'called', 'using', 'able', 'many', 'another', 'based', 'even', 'will', 'two', 'use', 'make', 'used', 'first', 'new'}
```

```python
all_stopwords = additional_stopwords.union(STOPWORDS)

if debug:
    print(all_stopwords)
```

```python
article = wikipedia.page(page)
if debug:
    print (article.url)
```

```python
wc = WordCloud(width=1000, height=1000, background_color="white", stopwords=all_stopwords).generate(article.content)
```

```python
plt.figure(figsize=(18, 18))
plt.axis("off")
plt.imshow(wc);
```
